```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv("./data/threads_reviews.csv")
```

```python
df.groupby("rating").size()
```

```python
print(df.shape)
```

```python
df.head(4)
```

```python
from wordcloud import WordCloud
text = ' '.join(df["review_description"])+" "
wordcloud = WordCloud(width = 800, height = 800,background_color ='white',min_font_size = 10).generate(text)
plt.figure(figsize = (20, 20), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0) 
plt.show()
```

```python
rate = df["rating"].value_counts().reset_index()
rate.columns = ['Rating', 'Count']
fig = px.bar(rate,x='Rating',y='Count', color='Rating', title='Rating Distribution')
fig.update_xaxes(title_text='Rating')
fig.update_yaxes(title_text='Count')
```

```python
count = df["source"].value_counts().reset_index()
count.columns = ['Source', 'Count']
fig = px.pie(count,names='Source', values='Count')
fig.show()
```
